```python
from PIL import Image
import numpy as np
import skimage.measure
import math
from matplotlib import pyplot as plt
```

```python
low_image = Image.open("X-Rays/70KV.jpg")
plt.imshow(low_image)
```

## EXTRACTING THE REGION OF INTEREST

```python
left = 1000
top = 500
right = 1050
bottom = 550
low_cropped_img = low_image.crop((left,top,right,bottom))
# low_cropped_img.show()
plt.imshow(low_cropped_img)
```

```python
np.unique(low_cropped_img)
```

## WIDTH AND HEIGHT OF ROI REGION

```python
w1,h1 = low_cropped_img.size
w1,h1 
```

### FINDING THE SUM OF PIXEL VALUES

```python
i_d = np.sum(low_cropped_img)
i_d
```

### CALIBRATION FACTOR

```python
width,i_h = low_image.size
w,r_h = low_cropped_img.size
c_f = i_h/r_h
c_f
```

### FINDING BONE MINERAL CONTENT

```python
bone_mineral_content = (i_d * 1.073)/c_f
bone_mineral_content
```

### FINDING DIAGONAL OF ROI

```python
diagonal = math.sqrt(pow(w1,2)+pow(h1,2))
```

```python
a = 3.14*(pow(diagonal/2,2))
```

```python
a
```

```python
entropy = skimage.measure.shannon_entropy(low_cropped_img)
```

```python
entropy
```

### FINDING VOLUME OF ROI

```python
volume_of_roi = a*entropy
```

```python
volume_of_roi
```

```python
bone_mineral_density = bone_mineral_content/volume_of_roi
```

```python
bone_mineral_density
```
